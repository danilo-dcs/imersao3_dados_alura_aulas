```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # importando o seaborn
sns.set()             # aplicando as configs do seaborn

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression='zip')
```

```python
dados.head()
```

```python
# realizando o cruzamento dos dados com o crosstable
pd.crosstab(dados['dose'],dados['tempo'])
```

```python
# realizando o cruzamento dos dados com o crosstable multiíndex
pd.crosstab([dados['dose'],dados['tempo']], dados['tratamento'])
```

```python
# realizando o cruzamento dos dados com o crosstable multiíndex
pd.crosstab([dados['dose'],dados['tempo']], dados['tratamento'], normalize=True)
```

```python
# realizando o cruzamento dos dados com o crosstable multiíndex, normalizando pela linha, ou seja, a soma dos itens da linha = 1
pd.crosstab([dados['dose'],dados['tempo']], dados['tratamento'], normalize='index' )
```

```python
# Analisar algum valor (expressão gênica), para ver a diferença de um determinado gene específico
# os valores exibidos dizem respeito às médias das expressões gênicas g-0 com droga e sem droga, tendo em visto um tempo e uma dose
pd.crosstab([dados['dose'],dados['tempo']], dados['tratamento'], values=dados['g- 0'], aggfunc='mean' )
```

```python
#gráfico de dispersão de dados, serve para encontrar padrões anaçisanso-se a variação de um parâmetro com relação à outro [g-0 e g-3]
sns.scatterplot(x='g-0', y='g-3', data=dados)
```

```python
#gráfico de dispersão de dados, serve para encontrar padrões anaçisanso-se a variação de um parâmetro com relação à outro
## é possível perceber que quanto mior o g-0, me nor é o g-8
sns.scatterplot(x='g-0', y='g-8', data=dados)
```

```python
#gráfico de expressão com linha de tendência, o parâmetro line_kws configura a linha
sns.lmplot(data=dados, x='g-0', y='g-8', line_kws={'color':'red'}, col='tratamento')
```

```python
#gráfico de expressão com linha de tendência, o parâmetro line_kws configura a linha
sns.lmplot(data=dados, x='g-0', y='g-8', line_kws={'color':'red'}, col='tratamento', row='tempo')
```

```python
# CORRELAÇÃO DOS GENES
# quando se tem uma correlação muito próxima de -1 ou 1, quer dizer que as variáveis estão muito relacionadas
# o sinal mais e menos indica se a correlação é direta oiu inversa
dados.loc[:,'g-0':'g-771'].corr()
```

```python
# Matriz de correlação com gráfico
import numpy as np

# Compute the correlation matrix
corr = dados.loc[:,'g-0':'g-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Compute the correlation matrix
corr_cel = dados.loc[:,'c-0':'c-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr_cel, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_cel, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
